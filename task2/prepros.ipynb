```python
import os
import cv2
import numpy as np
import warnings
from sklearn.model_selection import train_test_split
from random import shuffle
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.xception import preprocess_input
from tensorflow.keras.applications.vgg16 import VGG16
import tf2_yolov4
import matplotlib.pyplot as plt
from PIL import Image
from PIL import ImageFilter

warnings.filterwarnings("ignore")

IMG_SIZE = 50
```

```python
TRAIN_DIR = '/content/TrainImages'
TRAIN_BB_DIR = '/content/TrainGroundTruth'
TEST_DIR = '/content/TestImages'
TEST_BB_DIR = '/content/TestGroundTruth'

images = []
targets = []

for i in os.listdir(TEST_DIR):
    img = tf.keras.preprocessing.image.load_img(os.path.join(TRAIN_DIR, i), target_size=(IMG_SIZE, IMG_SIZE), color_mode='grayscale')
    img = np.asarray(img).astype('float32') / 255.0
    img = tf.keras.utils.img_to_array(img)
    img = np.array(img)
    images.append(img)
    target = i.split('.')[0]
    targets.append(target)
```
